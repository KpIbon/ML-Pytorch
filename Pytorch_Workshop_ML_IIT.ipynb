# PyTorch Fundamentals

In this section, we'll cover the core building blocks of PyTorch:
- **Tensors:** The fundamental data structure in PyTorch.
- **Tensor Operations:** Basic arithmetic, reshaping, slicing, and other manipulations.
- **Autograd:** PyTorch's automatic differentiation engine for computing gradients.

Let's start by understanding what tensors are and how to work with them.

image.png

```python
# Creating tensors in various ways:
# 1. Direct creation from a Python list
tensor_from_list = torch.tensor([1.0, 2.0, 3.0])
print("Tensor from list:", tensor_from_list)

# 2. Creating a tensor with uninitialized values (size: 3x3)
uninitialized_tensor = torch.empty(3, 3)
print("\nUninitialized tensor (3x3):\n", uninitialized_tensor)

# 3. Creating a tensor filled with zeros (size: 2x2)
zeros_tensor = torch.zeros(2, 2)
print("\nZeros tensor (2x2):\n", zeros_tensor)

# 4. Creating a tensor with random values (size: 2x2)
rand_tensor = torch.rand(2, 2)
print("\nRandom tensor (2x2):\n", rand_tensor)
```

### Basic Tensor Operations

Tensors support a wide range of operations similar to NumPy arrays. In this cell, we'll cover:
- **Arithmetic Operations:** Addition, subtraction, multiplication, and division.
- **Element-wise Operations:** Operations performed on each element.
- **Mathematical Functions:** Functions like exponentiation, logarithms, etc.

Let's see some examples in the next code cell.

```python
# Define two sample tensors for arithmetic operations
a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([4.0, 5.0, 6.0])

# Addition
add_result = a + b
print("Addition:", add_result)

# Subtraction
sub_result = a - b
print("Subtraction:", sub_result)

# Multiplication (element-wise)
mul_result = a * b
print("Element-wise Multiplication:", mul_result)

# Division (element-wise)
div_result = a / b
print("Element-wise Division:", div_result)

# Using built-in functions: exponentiation
exp_result = torch.exp(a)
print("Exponential:", exp_result)
```

### Tensor Manipulation

Tensors can be manipulated in various ways:
- **Indexing & Slicing:** Similar to Python lists and NumPy arrays.
- **Reshaping:** Changing the dimensions of a tensor without changing its data.

In the following code cell, we'll illustrate these operations with examples.

image.png

```python
# Creating a 4x4 tensor with values from 0 to 15
tensor_seq = torch.arange(16).reshape(4, 4)
print("Original 4x4 Tensor:\n", tensor_seq)

# Indexing: Get the element at row 2, column 3 (0-indexed)
element = tensor_seq[2, 3]
print("\nElement at row 2, column 3:", element)

# Slicing: Get the first two rows and the last two columns
sliced_tensor = tensor_seq[:2, 2:]
print("\nSliced Tensor (first 2 rows, last 2 columns):\n", sliced_tensor)

# Reshaping: Flatten the tensor to a 1D tensor
flattened_tensor = tensor_seq.view(-1)
print("\nFlattened Tensor:\n", flattened_tensor)
```

### Autograd: Automatic Differentiation

One of PyTorch's core features is **autograd**, which automatically computes gradients for tensor operations. This is crucial for training neural networks as it allows the framework to perform backpropagation easily.

- **requires_grad:** A flag that tells PyTorch to track operations on a tensor.
- **backward():** A function to compute gradients of a tensor with respect to some scalar value.

In the next cell, we'll see autograd in action.

```python
# Create a tensor and enable gradient tracking
x = torch.tensor([2.0, 3.0], requires_grad=True)
print("Tensor x:", x)

# Define a simple function: y = x^2 + 3x + 2
y = x**2 + 3*x + 2
print("\nTensor y (before summing):", y)

# For computing gradients, we need a scalar output.
# Sum the elements to obtain a scalar.
y_sum = y.sum()
print("\nScalar y_sum:", y_sum)

# Compute gradients: dy/dx
y_sum.backward()

# Display the gradients stored in x.grad
print("\nGradients dy/dx:", x.grad)
```

# Linear Regression Example

In this example, we'll build a linear regression model to predict a continuous target value. Weâ€™ll create synthetic data that follows a linear relationship (e.g., y = 2x + 1 with some noise) and use a single linear layer to learn this relationship.

The key steps are:
- Generating synthetic data.
- Defining a simple model with `nn.Linear`.
- Specifying the Mean Squared Error (MSE) loss function.
- Training the model using gradient descent.

image.png

```python
import matplotlib.pyplot as plt

# Set a seed for reproducibility
torch.manual_seed(42)

# Generate synthetic data: y = 2x + 1 with noise
x = torch.linspace(-10, 10, 100).reshape(-1, 1)
y = 2 * x + 1 + torch.randn(x.size()) * 2  # add some noise

# Visualize the data
plt.figure(figsize=(6,4))
plt.scatter(x.numpy(), y.numpy(), label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic Linear Regression Data')
plt.legend()
plt.show()
```

```python
import torch.nn as nn
import torch.optim as optim

# Define a simple linear regression model using nn.Linear
model_lr = nn.Linear(in_features=1, out_features=1)

# Define the loss function (Mean Squared Error) and the optimizer (SGD)
criterion_lr = nn.MSELoss()
optimizer_lr = optim.SGD(model_lr.parameters(), lr=0.001)

# Print initial model parameters
print("Initial model parameters:", list(model_lr.parameters()))
```

```python
# Training loop for linear regression
num_epochs = 10000
for epoch in range(num_epochs):
    model_lr.train()
    optimizer_lr.zero_grad()      # Clear gradients from the previous step

    # Forward pass: compute predictions
    predictions = model_lr(x)

    # Compute loss
    loss = criterion_lr(predictions, y)

    # Backward pass: compute gradients
    loss.backward()

    # Update parameters
    optimizer_lr.step()

    if (epoch+1) % 40 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# Visualize the fitted line against the data
predicted = model_lr(x).detach()
plt.figure(figsize=(6,4))
plt.scatter(x.numpy(), y.numpy(), label='Data Points')
plt.plot(x.numpy(), predicted.numpy(), color='red', label='Fitted Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression: Data and Fitted Line')
plt.legend()
plt.show()
```

# Simple Neural Network for Binary Classification

In this example, we'll create a simple neural network to classify points into two classes. We will:
- Generate synthetic 2D data for two classes.
- Build a neural network with one hidden layer.
- Use the CrossEntropyLoss (which combines a LogSoftmax and NLLLoss) to train our model.
- Train the model and print training loss to see how it learns to separate the classes.

Let's start by generating our synthetic dataset.

```python
import numpy as np

# Number of samples per class
n_samples = 100

# Generate synthetic data for class 0 and class 1
# Class 0: centered at (-2, -2)
x0 = torch.randn(n_samples, 2) + torch.tensor([-2, -2])
# Class 1: centered at (2, 2)
x1 = torch.randn(n_samples, 2) + torch.tensor([2, 2])

# Combine data and create labels
X = torch.cat([x0, x1], dim=0)
y_class = torch.cat([torch.zeros(n_samples, dtype=torch.long), torch.ones(n_samples, dtype=torch.long)])

# Visualize the dataset
plt.figure(figsize=(6,4))
plt.scatter(X[y_class==0, 0].numpy(), X[y_class==0, 1].numpy(), color='blue', label='Class 0')
plt.scatter(X[y_class==1, 0].numpy(), X[y_class==1, 1].numpy(), color='orange', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Synthetic Data for Binary Classification')
plt.legend()
plt.show()
```

```python
# Define a simple neural network with one hidden layer using nn.Module
class SimpleNN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out

# Create an instance of the neural network
input_dim = 2    # two features
hidden_dim = 15  # number of neurons in hidden layer
output_dim = 2   # two classes

model_nn = SimpleNN(input_dim, hidden_dim, output_dim)

# Print the model architecture
print(model_nn)
```

```python
# Define the loss function and optimizer for classification
criterion_nn = nn.CrossEntropyLoss()
optimizer_nn = optim.Adam(model_nn.parameters(), lr=0.01)

print("Using CrossEntropyLoss and Adam optimizer.")
```

```python
# Training loop for the neural network
num_epochs_nn = 500
for epoch in range(num_epochs_nn):
    model_nn.train()
    optimizer_nn.zero_grad()    # Reset gradients

    # Forward pass: compute predictions
    outputs = model_nn(X)

    # Compute loss (CrossEntropyLoss expects raw logits)
    loss = criterion_nn(outputs, y_class)

    # Backward pass: compute gradients and update weights
    loss.backward()
    optimizer_nn.step()

    if (epoch+1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs_nn}], Loss: {loss.item():.4f}')

# Optionally, evaluate the model on the training data
_, predicted = torch.max(model_nn(X), 1)
accuracy = (predicted == y_class).float().mean()
print(f'\nTraining Accuracy: {accuracy.item()*100:.2f}%')
```

```python
import numpy as np

# Create a meshgrid based on the feature space boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Prepare the grid data for prediction: combine xx and yy into a 2D array
grid = torch.tensor(np.c_[xx.ravel(), yy.ravel()], dtype=torch.float32)

# Set the model to evaluation mode and compute predictions for each grid point
model_nn.eval()
with torch.no_grad():
    outputs_grid = model_nn(grid)
    # Get predicted classes from logits
    _, predicted_grid = torch.max(outputs_grid, 1)

# Reshape the predictions to match the meshgrid shape
Z = predicted_grid.numpy().reshape(xx.shape)

# Plot the decision boundary using a contour plot
plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')

# Overlay the original data points with true labels
plt.scatter(X[y_class==0, 0].numpy(), X[y_class==0, 1].numpy(), color='blue', label='Class 0')
plt.scatter(X[y_class==1, 0].numpy(), X[y_class==1, 1].numpy(), color='orange', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundary for Neural Network Classification')
plt.legend()
plt.show()
```

```python
# Save the state dictionary of our neural network model
torch.save(model_nn.state_dict(), "simple_nn.pth")
print("Model saved to simple_nn.pth")
```

# Importing and Using Pre-built Powerful Models

In this section, we'll demonstrate how to leverage pre-built models in PyTorch. Weâ€™ll explore three examples:

1. **ResNet:** A pre-trained model for image classification.
2. **Transformer Model:** Using Hugging Faceâ€™s pipeline to perform sentiment analysis.
3. **Faster R-CNN:** A powerful object detection model from torchvision.

These examples show that you can easily import sophisticated models and use them for various tasks with just a few lines of code.

```python
import torch
from torchvision import models, transforms
from PIL import Image
import requests
import io
import json
import matplotlib.pyplot as plt

# Load a pre-trained ResNet-18 model and set it to evaluation mode
model_resnet = models.resnet18(pretrained=True)
model_resnet.eval()

# Download a sample image (a dog image)
# image_url = "https://cdn05.zipify.com/ncIhijRpIAr4_sSv1Ua5nb1XI6Y=/fit-in/3840x0/7066c6b08ea94801b47f43b44e7560a3/15.jpeg"
image_url = "https://www.vidavetcare.com/wp-content/uploads/sites/234/2022/04/golden-retriever-dog-breed-info.jpeg"
response = requests.get(image_url)
img = Image.open(io.BytesIO(response.content)).convert("RGB")

# Preprocess the image for ResNet: resize, center crop, convert to tensor, and normalize
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])
input_tensor = preprocess(img)
input_batch = input_tensor.unsqueeze(0)  # add batch dimension

# Run inference and get the predicted class index
with torch.no_grad():
    output = model_resnet(input_batch)
predicted_class = output.argmax(dim=1).item()
print(predicted_class)

# Retrieve human-readable labels for ImageNet classes
imagenet_url = "https://s3.amazonaws.com/deep-learning-models/image-models/imagenet_class_index.json"
response = requests.get(imagenet_url)
imagenet_classes = json.loads(response.content)
predicted_label = imagenet_classes[str(predicted_class)][1]
print("Predicted class index (ResNet):", predicted_class)
print("Predicted label (ResNet):", predicted_label)

# Visualize the image with a title showing the prediction
plt.figure(figsize=(6,4))
plt.imshow(img)
plt.title(f"ResNet Prediction: {predicted_label}")
plt.axis("off")
plt.show()
```

```python
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision import transforms
from PIL import Image, ImageDraw
import requests
import io
import matplotlib.pyplot as plt
import torch

# Load a pre-trained Faster R-CNN model for object detection and set it to evaluation mode
model_detection = fasterrcnn_resnet50_fpn(pretrained=True)
model_detection.eval()

# Use an image URL that clearly shows a dog (Golden Retriever image)
image_url = "https://www.catsanddogsla.com/uploads/4/7/8/9/47898741/187-4341x2402-300dpi-all-free-download-com-15484942_orig.jpg"
response = requests.get(image_url)
img_url = Image.open(io.BytesIO(response.content)).convert("RGB")

# Convert the image to a tensor (Faster R-CNN expects a list of tensors)
detection_transform = transforms.ToTensor()
detection_input = detection_transform(img_url)

# Run object detection (model expects a list of tensors as input)
with torch.no_grad():
    predictions = model_detection([detection_input])

# COCO category names list (index 17 corresponds to 'dog')
COCO_INSTANCE_CATEGORY_NAMES = [
    '__background__', 'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
    'train', 'truck', 'boat', 'traffic light', 'fire hydrant', 'stop sign',
    'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow',
    'elephant', 'bear', 'zebra', 'giraffe', 'backpack', 'umbrella', 'handbag',
    'tie', 'suitcase', 'frisbee', 'skis', 'snowboard', 'sports ball', 'kite',
    'baseball bat', 'baseball glove', 'skateboard', 'surfboard', 'tennis racket',
    'bottle', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl', 'banana',
    'apple', 'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza',
    'donut', 'cake', 'chair', 'couch', 'potted plant', 'bed', 'dining table',
    'toilet', 'tv', 'laptop', 'mouse', 'remote', 'keyboard', 'cell phone',
    'microwave', 'oven', 'toaster', 'sink', 'refrigerator', 'book', 'clock',
    'vase', 'scissors', 'teddy bear', 'hair drier', 'toothbrush'
]

# Set a threshold for detection confidence
threshold = 0.7

# Debug: Print out all predictions above threshold
print("High-confidence predictions:")
for box, label, score in zip(predictions[0]['boxes'], predictions[0]['labels'], predictions[0]['scores']):
    if score >= threshold:
        print(f"{COCO_INSTANCE_CATEGORY_NAMES[label]}: {score:.2f}")

# Draw bounding boxes on the image for detected 'dog'
draw = ImageDraw.Draw(img_url)
dog_detected = False
for box, label, score in zip(predictions[0]['boxes'], predictions[0]['labels'], predictions[0]['scores']):
    if score >= threshold and COCO_INSTANCE_CATEGORY_NAMES[label] == 'dog':
        box = box.tolist()
        draw.rectangle(box, outline="red", width=3)
        draw.text((box[0], box[1]), f"dog: {score:.2f}", fill="red")
        dog_detected = True

if dog_detected:
    print("Dog detected in the image!")
else:
    print("No dog detected with high confidence.")

# Display the image with detections
plt.figure(figsize=(8,6))
plt.imshow(img_url)
plt.title("Faster R-CNN: Dog Detection")
plt.axis('off')
plt.show()
```
